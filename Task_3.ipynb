# Task-03:  ‘Exploratory Data Analysis’ on dataset ‘SampleSuperstore’ 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df
```

# Exploratory Data Analysis

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
Top_Sales = df.groupby("State").Sales.sum().nlargest(n =15)
Top_Profit = df.groupby("State").Profit.sum().nlargest(n=15)
Top_Sales.index
```

```python
Top_Profit.index
```

```python
Top_Sales.plot(kind='bar',figsize=(14,8))
```

```python
Top_Profit.plot(kind='bar',figsize=(14,8))
```

We can see that Sales and Profit are directly related to eachother. If there are more sales then profit is more.

```python
sns.heatmap(df.corr(), annot=True)
```

```python
sns.catplot(data=df, kind='bar', x='Discount', y='Sales')
```

```python
sns.catplot(data=df, kind='bar', x='Discount', y='Profit')
```

From this we can see that if discount is given profit is less and sales is more.Which means that profit and discount are not directly proportional 

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(24,8))
plt.xticks(rotation=90)
sns.barplot(data=df, x = 'State', y='Profit')
```

```python
sns.countplot(data=df,x='Ship Mode')
```

```python
sns.catplot(data=df, kind='bar', x='Region', y='Sales')
```

```python
sns.catplot(data=df, kind='bar', x='Region', y='Profit')
```

```python
sns.catplot(data=df, kind='bar', x='Region', y='Discount')
```

```python
sns.pairplot(df,hue='Region')
```

```python
sns.countplot(data = df, x = 'Category',hue='Region')
```

* West region is high in buying in all the categories

```python
plt.figure(figsize=(20,10))
sns.countplot(data = df, x = 'Sub-Category',hue='Region')
plt.title('Sub-Category',fontsize=15)
```

```python
sns.countplot(data=df, x = 'Segment')
```

```python
sns.catplot(data=df, kind = 'bar',x = 'Category',y = 'Sales')
```

```python
sns.catplot(data=df, kind = 'bar',x = 'Category',y = 'Profit')
```

* Sales and profit are more in technology category
* Profit is less in Furniture Category

```python
plt.figure(figsize=(20,6))
sns.barplot(data=df, x='Sub-Category', y='Sales')
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data=df, x='Sub-Category', y='Profit')
```

* Subcategory Copiers is with high sale and profit
* Machines sales is high but profit is less
